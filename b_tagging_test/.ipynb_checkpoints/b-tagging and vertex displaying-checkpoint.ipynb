```python
from heppy.papas.path import *
from utils.computeIP import *
```

```python
jet_direction = TVector3(0, 1, 0.2)
pr_vertex = TVector3(0, 0, 0)

track_p4 = TLorentzVector()
track_p4.SetPxPyPzE(0.2, 0.6, 0, 1)
track_origin = TVector3(3e-5, 5e-5, 0)

magnetic_field = 3800
charge = 1
```

```python
track1 = Helix(magnetic_field, charge, track_p4, track_origin)
compute_IP(track1, pr_vertex, jet_direction)
v_displayer = vertex_displayer("trial_simple", "Trial with IP simple", 1e-4, -1e-12, 1e-12, track1, 100, 'simple')
v_displayer.draw()
v_displayer.c_transverse.Draw()
```

```python
track2 = Helix(magnetic_field, charge, track_p4, track_origin)
compute_IP_wrt_direction(track2, pr_vertex, jet_direction)
v_displayer = vertex_displayer("trial_complex", "Trial with IP complex", 1e-4, -1e-12, 1e-12, track2, 100, 'complex')
v_displayer.draw()
v_displayer.c_transverse.Draw()
```
